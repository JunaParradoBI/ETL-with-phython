```python
import pandas as pd
from sodapy import Socrata

# Cliente sin autenticar solo funciona con conjuntos de datos públicos.
client = Socrata("mydata.iadb.org", None)

# Identificador del conjunto de datos específico de SIMS (asegúrate de usar el correcto)
dataset_identifier = "v2c9-36h7"

# Parámetros de configuración
limit = 1000  # Número máximo de filas por solicitud
offset = 0  # Desplazamiento inicial
all_results = []

# Filtrar por año 2020
where_clause = "anio = 2020"

# Obtener datos en bloques usando paginación
while True:
    results = client.get(dataset_identifier, where=where_clause, limit=limit, offset=offset)
    if not results:  # Si no hay más resultados, salir del bucle
        break
    all_results.extend(results)
    offset += limit
    print(f"Obtenidos {offset} filas...")

# Convertir los resultados a un DataFrame de pandas
results_df = pd.DataFrame.from_records(all_results)

# Guardar el DataFrame en un archivo Excel
results_df.to_excel("resultados.xlsx", index=False)

# Mostrar las primeras filas del DataFrame para ver la estructura de los datos
print(results_df.head())

# results_df['some_column'] = pd.to_numeric(results_df['some_column'], errors='coerce')
```
